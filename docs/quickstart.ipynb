# bsts-causalimpact Quickstart

This notebook demonstrates the basic usage of bsts-causalimpact in under 5 minutes.

```bash
pip install bsts-causalimpact
```

````python
import numpy as np
import pandas as pd
from causal_impact import CausalImpact
````

## 1. Generate Synthetic Data

We create a time series where:
- `y` is the response variable (e.g., sales)
- `x1` is a covariate (e.g., ad spend in a control market)
- An intervention at day 71 causes a +3.0 lift in `y`

````python
rng = np.random.default_rng(42)
n = 100
dates = pd.date_range("2020-01-01", periods=n, freq="D")

x1 = rng.normal(5, 1, n)
y = 1.0 * x1 + rng.normal(0, 0.5, n)

# Add causal effect after day 70
y[70:] += 3.0

data = pd.DataFrame({"y": y, "x1": x1}, index=dates)
data.head()
````

## 2. Define Periods and Run Analysis

````python
pre_period = ["2020-01-01", "2020-03-11"]
post_period = ["2020-03-12", "2020-04-09"]

ci = CausalImpact(data, pre_period, post_period)
````

## 3. View Summary

````python
print(ci.summary())
````

## 4. View Report

A narrative interpretation of the results:

````python
print(ci.report())
````

## 5. Plot Results

Three panels: original series with counterfactual, pointwise effect, and cumulative effect.

````python
fig = ci.plot()
fig.set_size_inches(10, 8)
fig
````

## 6. Access Raw Results

For programmatic access to per-timestep effects and aggregate statistics:

````python
# Per-timestep inferences
ci.inferences.head()
````

````python
# Aggregate statistics
ci.summary_stats
````

## 7. Using ModelOptions (Typed Config)

Instead of passing a dict, you can use `ModelOptions` for validated configuration:

````python
from causal_impact import ModelOptions

opts = ModelOptions(niter=2000, nwarmup=1000, seed=42)
ci2 = CausalImpact(data, pre_period, post_period, model_args=opts)
print(ci2.summary())
````
